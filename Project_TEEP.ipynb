```python
import cv2
import os

video_file = '/Users/prathamraj/Desktop/video/Camera7_20211231235959.avi'
vid = cv2.VideoCapture(video_file)

frame_rate = vid.get(cv2.CAP_PROP_FPS)

frame_interval = int(frame_rate)

output_folder = 'lobby_1'
os.makedirs(output_folder, exist_ok=True)

cf = 0

while True:
    ret, frame = vid.read()
    
    if ret:
        if cf % frame_interval == 0:
            name = f'{output_folder}/frame_{cf}.jpg'
            print('Creating ' + name)
            cv2.imwrite(name, frame)
            
        cf += 1
    else:
        break

vid.release()
```

```python
import os
import cv2
import shutil
from ultralytics import YOLO

# Load the pre-trained YOLOv5 model for human detection
model = YOLO('yolov5su.pt')  # You can use 'yolov5m.pt', 'yolov5l.pt', etc., based on your accuracy and speed requirements

# Define the folder paths
input_folder_path = '/Users/prathamraj/Desktop/teep_pr/lobby_1'
output_folder_path = '/Users/prathamraj/Desktop/teep_pr/segre_lobby_1'

# Create the output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# Loop through each image in the input folder
for filename in os.listdir(input_folder_path):
    if filename.endswith('.jpg') or filename.endswith('.jpeg'):
        # Construct the full image path
        image_path = os.path.join(input_folder_path, filename)
        
        # Read the image
        image = cv2.imread(image_path)
        
        # Perform object detection
        results = model(image)
        
        # Check if any human is detected in the image
        human_detected = False
        for result in results:
            # Iterate through all the detected objects
            for obj in result.boxes:
                if int(obj.cls[0]) == 0:  # Class '0' corresponds to 'person' in COCO dataset used by YOLO
                    human_detected = True
                    break
        
        # If a human is detected, move the image to the output folder
        if human_detected:
            output_image_path = os.path.join(output_folder_path, filename)
            shutil.move(image_path, output_image_path)
            print(f'Moved {filename} to {output_folder_path} - Human detected')

print('Processing completed.')
```

```python
import os
import cv2
import shutil
from ultralytics import YOLO

# Load the pre-trained YOLOv5 model for human detection
model = YOLO('yolov5su.pt')  # You can use 'yolov5m.pt', 'yolov5l.pt', etc., based on your accuracy and speed requirements

# Define the folder paths
input_folder_path = 'Users/prathamraj/Desktop/teep_pr/segre_lobby_1'
output_base_folder_path = 'Users/prathamraj/Desktop/teep_pr/class_lobby_1'

# Create the base output folder if it doesn't exist
os.makedirs(output_base_folder_path, exist_ok=True)

# Initialize a dictionary to store person folders
person_folders = {}

# Initialize a person counter for unique identification
person_counter = 0

# Loop through each image in the input folder
for filename in os.listdir(input_folder_path):
    if filename.endswith('.jpg') or filename.endswith('.jpeg'):
        # Construct the full image path
        image_path = os.path.join(input_folder_path, filename)
        
        # Read the image
        image = cv2.imread(image_path)
        
        # Perform object detection
        results = model(image)

        for result in results:
            # Iterate through all the detected objects
            for obj in result.boxes:
                if int(obj.cls[0]) == 0:  # Class '0' corresponds to 'person' in COCO dataset used by YOLO
                    # Assign a unique ID for the detected person
                    person_counter += 1
                    
                    # Get the bounding box coordinates
                    x1, y1, x2, y2 = map(int, obj.xyxy[0])
                    
                    # Crop the detected person from the image
                    cropped_person = image[y1:y2, x1:x2]
                    
                    # Create a folder for the detected person if it doesn't exist
                    person_folder = os.path.join(output_base_folder_path, f'person_{person_counter}')
                    if not os.path.exists(person_folder):
                        os.makedirs(person_folder)
                        person_folders[person_counter] = person_folder
                    
                    # Save the cropped image to the person's folder
                    output_image_path = os.path.join(person_folder, f'{filename}')
                    cv2.imwrite(output_image_path, cropped_person)
                    print(f'Saved cropped image of person {person_counter} from {filename} to {output_image_path}')

print('Processing completed.')
```
